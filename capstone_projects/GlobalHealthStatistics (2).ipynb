```r
unique_diseases <- unique(data$Disease.Name)
num_unique_diseases <- length(unique_diseases)
print(paste("Total number of unique diseases:", num_unique_diseases))

library(ggplot2)

disease_counts <- as.data.frame(table(data$Disease.Name))
colnames(disease_counts) <- c("Disease.Name", "Frequency")

plot <- ggplot(disease_counts, aes(x = Disease.Name, y = Frequency)) +
  geom_bar(stat = "identity", fill = "skyblue") +
  labs(title = "Frequency of Unique Diseases",
       x = "Disease Name",
       y = "Frequency") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 90, hjust = 1, size = 12),
    axis.text.y = element_text(size = 12),
    axis.title = element_text(size = 14),
    plot.title = element_text(size = 16, face = "bold")
  )


print(plot)

ggsave("disease_frequency_plot.png", plot, width = 12, height = 6, dpi = 300)
```

```r
library(ggplot2)

# Create the plot
plot <- ggplot(data, aes(x = reorder(Country, Prevalence.Rate), y = Prevalence.Rate, fill = Disease.Name)) +
  geom_bar(stat = "identity") +
  labs(title = "Disease Prevalence Rate by Country",
       x = "Country",
       y = "Prevalence Rate (%)") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 90, hjust = 1, size = 12),  # Increase country label size
    axis.text.y = element_text(size = 12),                        # Increase y-axis text size
    axis.title = element_text(size = 14),                         # Increase axis title size
    plot.title = element_text(size = 18, face = "bold"),          # Increase title size
    strip.text = element_text(size = 14, face = "bold")           # Increase facet title size
  ) +
  facet_wrap(~ Disease.Name, scales = "free_y") +
  scale_fill_viridis_d()


print(plot)


ggsave("large_disease_prevalence_plot.png", plot, width = 20, height = 12, dpi = 300)
```

```r
if (any(data$Country == "India")) {
  print("India is present in the dataset.")
} else {
  print("India is not present in the dataset.")
}
```

```r

india_count <- sum(data$Country == "India")
other_count <- sum(data$Country != "India")

pie_data <- data.frame(
  Category = c("India", "Other Countries"),
  Count = c(india_count, other_count)
)

pie_data$Percentage <- pie_data$Count / sum(pie_data$Count) * 100

library(ggplot2)

ggplot(pie_data, aes(x = "", y = Count, fill = Category)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar("y", start = 0) +
  geom_text(aes(label = paste0(round(Percentage), "%")),
            position = position_stack(vjust = 0.5)) +
  labs(title = "Proportion of Data Related to India",
       fill = "Category") +
  theme_void()
```

```r

india_data <- subset(data, Country == "India")
india_data<-na.omit(india_data)
```

```r
head(india_data)
```

```r
num_rows <- nrow(india_data)
print(paste("Number of rows in india_data:", num_rows))
```

```r

ggplot(india_data, aes(x = reorder(Disease.Name, Prevalence.Rate), y = Prevalence.Rate)) +
  geom_col(fill = "steelblue") +
  labs(title = "Prevalence Rate of Diseases in India",
       x = "Disease Name",
       y = "Prevalence Rate") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))


if ("Year" %in% colnames(india_data)) {
  ggplot(india_data, aes(x = Year, y = Prevalence.Rate, group = 1)) +
    geom_line(color = "darkgreen") +
    geom_point(color = "darkgreen") +
    labs(title = "Prevalence Rate Trend in India Over Time",
         x = "Year",
         y = "Prevalence Rate")
} else {
  message("Year column not found in india_data. Cannot create a time-series plot.")
}


if ("Disease.Category" %in% colnames(india_data)) {
  ggplot(india_data, aes(x = Disease.Category, y = Prevalence.Rate, fill = Disease.Category)) +
    geom_boxplot() +
    labs(title = "Distribution of Prevalence Rate by Disease Category in India",
         x = "Disease Category",
         y = "Prevalence Rate") +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
} else {
  message("Disease.Category column not found in india_data. Cannot create a boxplot by category.")
}
```
